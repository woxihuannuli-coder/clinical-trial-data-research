```python
df_clean["start_date"] = pd.to_datetime(df_clean["start_date"],errors="coerce")
df_clean["completion_date"] = pd.to_datetime(df_clean["completion_date"],errors="coerce")

df_clean[["start_date","completion_date"]].head()
```

```python
object_cols = df_clean.select_dtypes(include ="object").columns
df_clean[object_cols] = df_clean[object_cols].apply(lambda col: col.str.strip())
df_clean[object_cols] = df_clean[object_cols].replace("", pd.NA)

for col in object_cols:
    print(f"\n===== {col} =====")
    print(df_clean[col].value_counts(dropna=False).head(10))
```

```python
df_clean.to_csv("../data/cleaned_trials.csv", index=False, encoding="utf-8-sig")
```
