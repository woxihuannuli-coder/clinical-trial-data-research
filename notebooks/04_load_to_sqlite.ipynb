```python
import pandas as pd
```

```python
import sqlite3
```

```python
df = pd.read_csv("../data/cleaned_trials.csv")
```

```python
conn = sqlite3.connect("../data/clinical_trials.db")
```

```python
df.to_sql(
    "clinical_trials",
    conn,
    if_exists="replace",
    index = False
)

conn.close()
```

```python
conn = sqlite3.connect("../data/clinical_trials.db")

query = """
SELECT COUNT(*) AS total_rows
FROM clinical_trials;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
PRAGMA table_info(clinical_trials);
"""

pd.read_sql_query(query, conn)
```
