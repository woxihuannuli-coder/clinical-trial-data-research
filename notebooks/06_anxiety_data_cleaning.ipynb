```python
import pandas as pd
```

```python
path = "../data/raw_anxiety_trials.csv"
```

```python
anxiety_df = pd.read_csv(path)
```

```python
anxiety_df.shape
```

```python
anxiety_df.head()
```

```python
anxiety_df.columns.tolist()
```

```python
# 컬럼명을 snake_case로 변경
```

```python
selected_colunms =["NCT Number",
    "Study Title",
    "Study Status",
    "Conditions",
    "Interventions",
    "Sponsor",
    "Phases",
    "Locations",
    "Start Date",
    "Completion Date"]

anxiety_base = anxiety_df[selected_colunms].copy()
anxiety_base.head()
```

```python
anxiety_base = anxiety_base.rename(columns={
    "NCT Number":"nct_id",
    "Study Title":"study_title",
    "Study Status":"study_status",
    "Conditions": "conditions",
    "Interventions":"interventions",
    "Sponsor":"sponsor",
    "Phases":"phases",
    "Locations":"locations",
    "Start Date":"start_date",
    "Completion Date":"completion_date"
        })
anxiety_base.head()
```

```python
anxiety_base["start_date"] = pd.to_datetime(anxiety_base["start_date"], errors="coerce")
anxiety_base["completion_date"] = pd.to_datetime(anxiety_base["completion_date"], errors="coerce")

anxiety_base.dtypes
```

```python
text_colunms = ['nct_id',
 'study_title',
 'study_status',
 'conditions',
 'interventions',
 'sponsor',
 'phases',
 'locations'
            ]

for col in text_colunms:
    anxiety_base[col] = anxiety_base[col].astype("string").str.strip()
```

```python
anxiety_base = anxiety_base.replace("",pd.NA)
```

```python
anxiety_base.head()
```

```python
before_count = len(anxiety_base)
cleaned_anxiety = anxiety_base.drop_duplicates(subset="nct_id").copy()
```

```python
after_count = len(cleaned_anxiety)
```

```python
print("before_count",before_count)
print("after_count",after_count)
```

```python
cleaned_anxiety["phases"] = cleaned_anxiety["phases"].fillna("NO_PHASE")
cleaned_anxiety["phases"].value_counts()
```

```python
save_path = "../data/cleaned_anxiety_trials.csv"
cleaned_anxiety.to_csv(save_path, index=False)
print("저장완료",save_path)
```
