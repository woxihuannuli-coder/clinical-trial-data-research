```python
import pandas as pd
import sqlite3
```

```python
depression_df = pd.read_csv("../data/cleaned_trials.csv")
anxiety_df = pd.read_csv("../data/cleaned_anxiety_trials.csv")

print("Depression",depression_df.shape)
print("Anxiety",anxiety_df.shape)
```

```python
depression_df["condition_group"] = "Depression"
anxiety_df["condition_group"]="Anxiety"
```

```python
print(depression_df.columns.tolist())
```

```python
print(anxiety_df.columns.tolist())
```

```python
depression_df = depression_df.rename(columns={
    "Locations": "locations"
})
```

```python
depression_df.columns.tolist()
```

```python
mental_health_df = pd.concat(
    [depression_df, anxiety_df],
    ignore_index= True)

mental_health_df.shape
```

```python
mental_health_df.columns.tolist()
```

```python
import sqlite3

db_path = "../data/mental_health_trials.db"

conn = sqlite3.connect(db_path)

depression_df.to_sql(
    "depression_trials",
    conn,
    if_exists="replace",
    index=False
)

anxiety_df.to_sql(
    "anxiety_trials",
    conn,
    if_exists="replace",
    index=False
)

mental_health_df.to_sql(
    "mental_health_trials",
    conn,
    if_exists="replace",
    index=False
)

conn.close()

print("DB 저장 완료:", db_path)
```

```python
conn = sqlite3.connect("../data/mental_health_trials.db")

check = pd.read_sql("""
SELECT condition_group, COUNT(*) AS trial_count
FROM mental_health_trials
GROUP BY condition_group;
""", conn)

conn.close()

check
```
